# Our implemation of Sampling

```python
import pandas as pd
import numpy as np
```

```python
rain = pd.read_csv('./data/derekswork/DTW_prec.csv')
```

```python
rain #That's a lot of data! If it was gigabytes, we could use just a sample for testing while developing.
```

```python
sample = rain.sample(n=5)
sample
#Different sample each time this cell is run (sampled without replacement)
```

```python
sample_based_on_percent = rain.sample(frac=0.1, random_state = 69)
sample_based_on_percent
```

#### notice the above cell always has 619 values (if frac=0.1). As in, there's no variance, so this is more deterministic than a true random binomial sample.

#### Sometimes it's in our interest to sample WITH replacement, allowing a data point to be taken more than once through multiple draws, without being exhausted. This is how dice work, for example. (rolling a 1 doesn't exclude a 1 from being rolled next) 

#### In such a large dataset as this, the difference between the two methods is small, but still there.

```python
sample_with_replacement = rain.sample(frac=0.1, replace=True, random_state=69)
sample_with_replacement
```
