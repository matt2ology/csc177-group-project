# Our implementation of Discretization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### in this example of discretization, we check winning lottery numbers from 2010 and later. 

#### Choose 5 numbers between 1 and 69, and one number between 1 and 26. This last number is the "powerball".
#### In order to win the jackpot, you must match every number.

```python
powerball = pd.read_csv("./data/derekswork/Lottery_Powerball_Winning_Numbers__Beginning_2010.csv")
```

```python
powerball.head()
```

```python
columnOfWinNum = powerball['Winning Numbers']
```

```python
type(columnOfWinNum)
```

```python
columnOfWinNum.head()
```

```python
columnOfWinNum[0]
```

```python
len(columnOfWinNum)
```

```python
columnOfWinNum[1][-2:]  #demonstrating a method to isolate the "powerball" number as a string
```

```python
listOfWinningPowerballNumbers = [] #putting all those powerballs into a list
```

```python
for i in range(len(columnOfWinNum)):
    str = columnOfWinNum[i][-2:]
    listOfWinningPowerballNumbers.append(int(str))
```

```python
listOfWinningPowerballNumbers #notice how some powerballs are above 26, even though it is impossible to win the jackpot this way
```

```python
dfOfWinningPowerballNumbers = pd.DataFrame(listOfWinningPowerballNumbers, columns=['num'])
```

```python
dfOfWinningPowerballNumbers
```

```python
dfOfWinningPowerballNumbers['num'].hist(bins=10)
```

### Note the dropoff in powerball numbers above 26. This is because the jackpot requires a powerball between 1 and 26. Some smaller prizes may be won, however, with a powerball above 26. (I think it is strictly less optimal to play with the powerball being greater than 26, but people clearly still do it. Maybe they don't read the rules before playing.)

#### Without discretization, this graph would be ugly, with many small bars, one for each number.
